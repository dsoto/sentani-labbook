# Interactive Graphs with Netatmo Data

The Bokeh library (which can be installed using conda) allows for interaction with the graphs.  Once you have the library installed, the code below should allow you to pan and zoom a plot which is great for data exploration.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
data = pd.read_csv('Main Room_1-5-2015.csv', 
                   skiprows=3,       # ignore the first 3 rows of data
                   sep=';',          # semicolon is used to separate data values
                   index_col=1,      # use column 1 as the dates to index the data
                   parse_dates=True) # convert the date string into a date object
```

```python
from bokeh.plotting import figure, output_notebook, show
output_notebook()
p = figure(x_axis_type='datetime')
p.line(y=data['CO2'], x=data.index)
show(p)
```
