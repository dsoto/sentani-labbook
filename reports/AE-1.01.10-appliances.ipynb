# Overall responses

I have plotted the percentage of households reporting ownership of the appliances in the survey.  These are expressed in two different ways.  The first assumes that a null response is a no.  The second omits the non-responses which is equivalent to assuming that the fraction of ownership in the null responses is equal to the responses.

```python
columns = ['app_now/lighting', 'app_now/TV', 'app_now/radio',
           'app_now/fridge', 'app_now/fan', 'app_now/rice_cooker',
           'app_now/welder'] 

# fillna(0) counts nulls as negative responses
appliance_fraction = survey[columns].fillna(0).mean()
appliance_fraction.sort(ascending=False)
appliance_fraction.plot(kind='bar', title='Fraction of Households Responding Yes (Counts Non-Responses as No)')
```

```python
appliance_fraction = survey[columns].mean()
appliance_fraction.sort(ascending=False)
appliance_fraction.plot(kind='bar', title='Fraction of Households Responding Yes (Omits Non-Responses)')
```

```python
keys = ['app_now/lighting', 'app_now/TV', 'app_now/radio',
                'app_now/fridge', 'app_now/fan', 'app_now/rice_cooker',
                'app_now/welder']

# function to aggregate responses for several columns into a dataframe
# can this be done with get_dummies?
def aggregate_responses(keys):
    dfdict = {}
    for col in keys:
        #result = survey[col].fillna('No Response').value_counts(dropna=True)
        result = survey[col].fillna('No Response').value_counts(dropna=False)
        result = result.rename(index={1.0:'Yes', 0.0:'No'})
        dfdict[col] = result
    aggregated = pd.DataFrame.from_dict(dfdict, orient='index')
    aggregated = aggregated[['Yes', 'No', 'No Response']]
    return aggregated

aggregated = aggregate_responses(keys)
aggregated.plot(kind='barh', stacked=True)
plt.show()
```

# Visualizations of correlations

These are a few examples of more and less correlated appliance ownership pairs.  These graphs probably aren't useful for publication but can be a useful visualization of how many households own two appliances.  The lower-left cluster shows how many households report zero (non-ownership) for both appliances and the upper-right cluster shows households reporting one (ownership) for both appliances.  The upper-left and lower-right show only ownership of one of the two.  The line is a regression line.  A line close to a 45 degree slope has a higher correlation than a flatter line.  The reason the points are in blocks is because they are 'jittered' by adding a random number to them so that they don't overlap and obscure their numbers. 

```python
sns.regplot('app_now/TV', 'app_now/lighting', data=survey, x_jitter=0.2, y_jitter=0.2)
plt.show()
```

# Conditional probability

The probability that a responding household has a refrigerator in this sample is about 19%.  If that household has a rice cooker, that probability goes up to 59%.  These high power loads are much more predictive of other high power loads.

```python
# returns the probability of the prior field given the evidence field
# also reports the naive probability
def conditional_probability(prior, evidence):
    prior_prob = survey[prior].mean()
    cond_prob = survey.groupby(evidence)[prior].mean()[1]
    print('probability of observing {} {:.3f}'.format(prior, prior_prob))
    print('probability of observing {} given {} {:.3f}'.format(prior, evidence, cond_prob))
```

```python
conditional_probability('app_now/fridge', 'app_now/rice_cooker')
conditional_probability('app_now/fridge', 'app_now/fan')
conditional_probability('app_now/rice_cooker', 'app_now/fan')
conditional_probability('app_now/rice_cooker', 'app_now/fridge')
conditional_probability('app_now/fan', 'app_now/fridge')
conditional_probability('app_now/fan', 'app_now/rice_cooker')
```

Lighting, television, and radio are not very predictive of other appliances.  That is, the probability of having one knowing that you have the other isn't very different.

```python
conditional_probability('app_now/lighting', 'app_now/TV')
conditional_probability('app_now/TV', 'app_now/lighting')
conditional_probability('app_now/TV', 'app_now/radio')
```
