# Key variables

This notebook looks at 

- Appliance ownership (cooking, lighting, cooling, communication)
- Energy Expenditure
- Income

according to village.  I have grouped the villages by access type to facilitate comparisons within one access category.

```python
# create derived columns
# adjust HP to daily assuming charging frequency is reported per week
survey['HP_daily'] = survey['HP_charging_price'] * survey['HP_charging_frequency'] * 52 / 365.25
survey['HP_monthly'] = survey['HP_charging_price'] * survey['HP_charging_frequency'] * 52 / 12

# adjust lighting to daily assuming reporting per week
survey['lighting_daily'] = survey['non_electrice_light_expenditures'] * 52 / 365.25
survey['lighting_monthly'] = survey['non_electrice_light_expenditures'] * 52 / 12

# generate more columns
survey['genset_monthly'] = survey['genset_expenditure'] * 30
survey['income_monthly'] = sti.income_monthly(survey)

survey['meta_village'] = sti.meta_village(survey)
survey['access_type'] = sti.access_type(survey)
```

We see that some villages report access to fans and other appliances despite a lack of grid access.  We could further investigate to see if these households own gensets.

```python
apps = survey.groupby(['access_type', 'meta_village'])[['app_now/TV', 
                                      'app_now/fridge',
                                      'app_now/fan',
                                      'app_now/rice_cooker']].mean()
apps.plot(kind='barh', figsize=(10,10))
plt.show()
```
