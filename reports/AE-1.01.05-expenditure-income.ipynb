The Consultant should search for and report any statistically significant correlations between specific energy expenditure variables and specific income indicator variables, as well as between any meta variables.  High priority.

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import drs_sentani as sti
survey = sti.get_survey()
```

The key question is, are there easily observable assets (besides appliances) that predict the level energy expenditure?

I haven't been able put significant effort into this yet.  I have one example of wall type, but I'd like to extend the analysis as time allows into other factors such as livestock.

```python
# create derived columns for monthly expenditures
survey['PLN_expenditure_monthly'] = survey['PLN_expenditure']
survey['community_microgrid_expenditure_monthly'] = survey['community_microgrid_expenditure']
survey['genset_expenditure_monthly'] = survey['genset_expenditure'] * 365.25 / 12

# adjust HP to monthly assuming charging frequency is reported per week
survey['HP_monthly'] = survey['HP_charging_price'] * survey['HP_charging_frequency'] * 52 / 12

# adjust lighting to monthly assuming reporting per week
survey['lighting_monthly'] = survey['non_electrice_light_expenditures'] * 52 / 12

# create access type column
survey['access_type'] = sti.access_type(survey)

temp = survey[['PLN_expenditure_monthly', 
               'community_microgrid_expenditure_monthly', 
               'genset_expenditure_monthly',
               'HP_monthly',
               'lighting_monthly']]
survey['energy_meta_monthly'] = temp.sum(axis=1)
```

```python
# average electricity expenditure by wall type?
#sns.boxplot
ax = sns.boxplot(survey['energy_meta_monthly'], groupby=survey['wall'])
ax.set_ylim((0, 4e6))
plt.show()
```

# Variation with construction type

When we look at all the villages, stone (batu) walls appear to have a median and 75% quantile above the other wall types.  When we further select by access type, other differences emerge.

```python
ax = sns.boxplot(survey['energy_meta_monthly'], groupby=survey['wall'])
ax.set_ylim((0, 4e6))
plt.show()
```

In areas with no electricity access, households with stone and wood walls have higher electricity expenditures than sago homes.

```python
temp = survey[(survey['access_type']=='no_access') & 
              (survey['demand_point']=='household')]
ax = sns.boxplot(temp['energy_meta_monthly'], groupby=survey['wall'])
ax.set_ylim((0, 4e6))
```

In grid locations, the differences between stone or wood and sago do not appear as pronounced.  Note that I have truncated some outliers above.

```python
temp = survey[(survey['access_type']=='PLN_grid') & 
              (survey['demand_point']=='household')]
ax = sns.boxplot(temp['energy_meta_monthly'], groupby=survey['wall'])
ax.set_ylim((0, 5e5))
```
