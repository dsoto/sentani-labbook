# Lighting

The majority of households have some type of lighting and report using it daily for seven hours each day.

The dominant source of lighting is electricity.  There do not seem to be any solar sources of lighting in the survey.

Kampung Baru and Yokiwa have a strong majority of kerosene for their lighting.

```python
survey['app_now/lighting'].value_counts(dropna=False)
```

```python
survey['app_lighting_per_wk'].median()
```

```python
survey['app_lighting_hrs'].median()
```

```python
# create subset of survey
keys = ['village_name',
        'lighting']
sdf = survey[keys]
# convert categorical to dummy columns
booled = pd.get_dummies(sdf, columns=['lighting'])
# find percentage of respondents by village
results = booled.groupby('village_name').mean()
ax = results.plot(kind='barh', stacked=True)
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

The list below is the most frequent response for each village for the type of lighting.

```python
results.idxmax(axis=1).str[9:]
```
