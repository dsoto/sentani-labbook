```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
import drs_sentani as sti
survey = sti.get_survey()
```

# Uniform frequency reporting

The PLN and microgrid questions asked for monthly values while the genset_expediture asked for weekly values.  The following code adjusts for these values and creates new temporary columns with the monthly values.  The PLN and microgrid expenditures are monthly, while the genset expenditure is daily.  Extrapolating the genset spending to a monthly cost will depend on whether the genset is the primary or backup power source.  This analysis assumes that the generator is used each day.

```python
frequency = {'PLN_expenditure':'monthly',
     'community_microgrid_expenditure':'monthly',
     'genset_expenditure':'daily'}

multiplier = {'monthly':1, 'weekly':4, 'daily':30}

# create new columns with same monthly frequency
for column in frequency.keys():
    # adjust expenditure frequency
    expenditure = survey[column] * multiplier[frequency[column]]
    new_column = column + '_monthly'
    survey[new_column] = expenditure
```

I break these out by village and access type.  Households in villages without community microgrid access show community microgrid expenditures but I'm not certain why.

```python
# create column for type of AC supply
survey['access_type'] = survey['village_name'].apply(lambda x: sti.access_map.get(x))
# calculate means
survey.groupby(['access_type', 
                'village_name'])[['PLN_expenditure_monthly',
                                  'genset_expenditure_monthly', 
                                  'community_microgrid_expenditure_monthly']].agg(['mean','count'])
```

## Genset delivered energy

Assuming 10,000 Rp/liter and 0.4 liter/kWh, a 40,000 Rp/day genset expenditure implies 

40,000 Rp/day / 10,000 Rp/liter / 0.4 liter/kWh = 10 kWh per day

for an efficient generator.

This is a substantial load.  These generators may be much less efficient (0.8 liter/kWh ?) but this is still a 5 kWh daily load.

## PLN grid delivered energy

PLN is a monthly expenditure and is about 5000 Rp/day.  This implies that grid consumption is around 1-5 kWh if tariffs range between 5000 Rp/kWh and 1000 Rp/kWh.

## Community microgrid

I'm not sure what these tariffs are in order to estimate electricity usage.

```python
dfd = {}
columns = ['PLN_expenditure_monthly',
           'community_microgrid_expenditure_monthly',
           'genset_expenditure_monthly']

for col in columns:
    dfd[col] = survey[col].describe()
pd.set_option('precision', 2)
pd.DataFrame(dfd).T
```

# Monthly overall electricity spending by village

This table presents the overall electricity spending by village and access type.  (This is one possibility towards an energy meta-variable but still needs consideration about how we are averaging.)

I can restrict the survey (but haven't here) by household demand points since there are some other demand points with large expenditures that skew the means.  

```python
survey['electricity_monthly'] = survey[['PLN_expenditure_monthly', 
                                        'community_microgrid_expenditure_monthly', 
                                        'genset_expenditure_monthly']].sum(axis=1)
#subsurvey = survey[survey['demand_point'] == 'household']
stats = survey.groupby(['access_type', 'village_name'])['electricity_monthly'].describe()
pd.set_option('precision', 2)
pd.DataFrame(stats).unstack()
```

# Genset expenditures and any microgrid access

There is only a small difference in the means between the genset expenditures with and without microgrid access.

I create a temporary column that is true if either PLN or community microgrid working is true.

```python
# create microgrid column
# there should be 55 records here in the True column
survey['power_supply_working/any_microgrid'] = (
    survey['power_supply_working/PLN_microgrid'].fillna(0).astype(bool)) | (
        survey['power_supply_working/community_microgrid'].fillna(0).astype(bool))
stats = survey.groupby('power_supply_working/any_microgrid')['genset_expenditure'].describe()
pd.set_option('precision', 2)
pd.DataFrame(stats).unstack()
```

# Genset expenditure and PLN grid access

```python
stats = survey.groupby('power_supply_working/PLN_grid')['genset_expenditure'].describe()
pd.set_option('precision', 2)
pd.DataFrame(stats).unstack()
```
